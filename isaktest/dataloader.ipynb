```python
import os
import matplotlib.pyplot as plt
import random
from dotenv import load_dotenv

load_dotenv()

dataset_path = os.getenv("DATA_PATH") + random.choice(["0", "2", "3"])

random_img = random.choice(os.listdir(dataset_path))

img = os.path.join(dataset_path, random_img)

plt.imshow(plt.imread(img))
```

```python
from ultralytics import YOLO

model = YOLO("yolo11n-pose.pt")
```

```python
results = model(img)

# plt.imshow(results[0])
results
```
