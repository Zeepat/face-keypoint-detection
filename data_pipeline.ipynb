```python
import sqlite3
import pandas
```

```python
conn = sqlite3.connect('data/AFLW/aflw.sqlite')

# Create a cursor object
cursor = conn.cursor()

# Execute a query to fetch all table names
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")

# Fetch all results
tables = cursor.fetchall()

# Print the table names
for table in tables:
    print(table[0])

# Close the connection
conn.close()
```

```python
conn = sqlite3.connect('data/AFLW/aflw.sqlite')

# Create a cursor object
cursor = conn.cursor()

# Execute a query to fetch all data from the FaceMetaData table
cursor.execute("SELECT * FROM FaceMetaData")

# Fetch all results
annotations = cursor.fetchall()

# Get the column names from the cursor description
column_names = [description[0] for description in cursor.description]

# Print the column names
print("Column Names:")
print(column_names)

df_FaceMetaData = pandas.DataFrame(annotations, columns=column_names)

# Close the connection
conn.close()

df_FaceMetaData.head()
```

```python
conn = sqlite3.connect('data/AFLW/aflw.sqlite')

# Create a cursor object
cursor = conn.cursor()

# Execute a query to fetch all table names
cursor.execute("SELECT * FROM FeatureCoords")

# Fetch all results
annotations = cursor.fetchall()

df_feature_coords = pandas.DataFrame(annotations, columns=['face_id', 'feature_id', 'x', 'y', 'idk'])

# Close the connection
conn.close()
```

```python
conn = sqlite3.connect('data/AFLW/aflw.sqlite')

# Create a cursor object
cursor = conn.cursor()

# Execute a query to fetch all table names
cursor.execute("SELECT * FROM Faces")

# Fetch all results
annotations = cursor.fetchall()

df_faces = pandas.DataFrame(annotations, columns=['face_id', 'file_id', 'db_id'])

# Close the connection
conn.close()
```

```python
conn = sqlite3.connect('data/AFLW/aflw.sqlite')

# Create a cursor object
cursor = conn.cursor()

# Execute a query to fetch all table names
cursor.execute("SELECT * FROM FeatureCoordTypes")

# Fetch all results
annotations = cursor.fetchall()

df_feature_coord_types = pandas.DataFrame(annotations, columns=['feature_id', 'feature_name', 'feature_code', '1', '2', '3'])

# Close the connection
conn.close()
```

```python
# Filter the dataframe to keep only the rows with feature_id 8, 11, 15, or 19
df_feature_coords_filtered = df_feature_coords[df_feature_coords['feature_id'].isin([8, 11, 15, 19])]

# Display the filtered dataframe
df_feature_coords_filtered
```

```python
# Pivot the dataframe
df_pivot = df_feature_coords_filtered.pivot(index='face_id', columns='feature_id', values=['x', 'y'])

# Flatten the multi-index columns
df_pivot.columns = [f'{feature}_{axis}' for axis, feature in df_pivot.columns]

# Display the resulting dataframe
df_pivot.head()
```

```python
df_pivot.dropna(inplace=True)
```

```python
df_pivot.info()
```

```python
df_pivot.head()
```

```python
cols = list(df_pivot.columns)

cols.insert(1, cols.pop(cols.index('8_y')))
cols.insert(3, cols.pop(cols.index('11_y')))
cols.insert(5, cols.pop(cols.index('15_y')))

final_df = df_pivot[cols]

final_df.head()
```

```python
final_df = final_df.merge(df_faces[['face_id', 'file_id']], on='face_id', how='left')
final_df.head()
```

```python
final_df.rename(columns={'8_x': 'LeftEye_x', '8_y': 'LeftEye_y', '11_x': 'RightEye_x', '11_y': 'RightEye_y', '15_x': 'Nose_x', '15_y': 'Nose_y', '19_x': 'Mouth_x', '19_y': 'Mouth_y'}, inplace=True)
final_df.head()
```

```python
final_df = final_df.merge(df_FaceMetaData[['face_id', 'sex']], on='face_id', how='left')
final_df.head()
```
