```python
conn = sqlite3.connect(database)

# Create a cursor object
cursor = conn.cursor()

# Execute a query to fetch all table names
cursor.execute("SELECT * FROM Faces")

# Fetch all results
annotations = cursor.fetchall()

df_faces = pandas.DataFrame(annotations, columns=['face_id', 'file_id', 'db_id'])

# Close the connection
conn.close()
```

```python
conn = sqlite3.connect(database)

# Create a cursor object
cursor = conn.cursor()

# Execute a query to fetch all table names
cursor.execute("SELECT * FROM FeatureCoordTypes")

# Fetch all results
annotations = cursor.fetchall()

df_feature_coord_types = pandas.DataFrame(annotations, columns=['feature_id', 'feature_name', 'feature_code', '1', '2', '3'])

# Close the connection
conn.close()
```

```python
# Filter the dataframe to keep only the rows with feature_id 8, 11, 15, or 19
df_feature_coords_filtered = df_feature_coords[df_feature_coords['feature_id'].isin([8, 11, 15, 19])]

# Display the filtered dataframe
df_feature_coords_filtered
```

```python
# Pivot the dataframe
df_pivot = df_feature_coords_filtered.pivot(index='face_id', columns='feature_id', values=['x', 'y'])

# Flatten the multi-index columns
df_pivot.columns = [f'{feature}_{axis}' for axis, feature in df_pivot.columns]

# Display the resulting dataframe
df_pivot.head()
```

```python
df_pivot.dropna(inplace=True)
```

```python
df_pivot.info()
```

```python
df_pivot.head()
```

```python
cols = list(df_pivot.columns)

cols.insert(1, cols.pop(cols.index('8_y')))
cols.insert(3, cols.pop(cols.index('11_y')))
cols.insert(5, cols.pop(cols.index('15_y')))

final_df = df_pivot[cols]

final_df.head()
```

```python
final_df = final_df.merge(df_faces[['face_id', 'file_id']], on='face_id', how='left')
final_df.head()
```

```python
final_df.rename(columns={'8_x': 'LeftEye_x', '8_y': 'LeftEye_y', '11_x': 'RightEye_x', '11_y': 'RightEye_y', '15_x': 'Nose_x', '15_y': 'Nose_y', '19_x': 'Mouth_x', '19_y': 'Mouth_y'}, inplace=True)
final_df.head()
```

```python
final_df = final_df.merge(df_FaceMetaData[['face_id', 'sex']], on='face_id', how='left')
final_df.head()
```

```python
final_df = final_df.merge(df_FaceRect[['face_id', 'X_box', 'Y_box', 'W_box', 'H_box']], on='face_id', how='left')
final_df.head()
```

```python
existing_files = set(os.listdir(IMAGE_DIR))
print(f"Total existing images found: {len(existing_files)}")
```

```python
missing_files = final_df[~final_df['file_id'].isin(existing_files)]
print(f"Total missing images: {len(missing_files)}")
```

```python
if not missing_files.empty:
    missing_files[['file_id']].to_csv(MISSING_IMAGES_LOG, index=False)
    print(f"List of missing images saved to {MISSING_IMAGES_LOG}")
```

```python
final_df_cleaned = final_df[final_df['file_id'].isin(existing_files)].reset_index(drop=True)
print(f"Cleaned annotations size: {len(final_df_cleaned)}")
```

```python
annotated_files = set(final_df_cleaned['file_id'])
all_existing_files = set(os.listdir(IMAGE_DIR))
unannotated_files = all_existing_files - annotated_files
print(f"Total unannotated images to remove: {len(unannotated_files)}")
```

```python
if unannotated_files:
    with open(UNANNOTATED_IMAGES_LOG, 'w') as log:
        for file in unannotated_files:
            log.write(f"{file}\n")
    print(f"List of unannotated images saved to {UNANNOTATED_IMAGES_LOG}")

    # **Safety Step**: Verify the list before deletion
    confirmation = input(f"Are you sure you want to delete {len(unannotated_files)} unannotated images? (yes/no): ")
    if confirmation.lower() == 'yes':
        # Remove unannotated images
        for file in unannotated_files:
            file_path = os.path.join(IMAGE_DIR, file)
            try:
                os.remove(file_path)
                logging.info(f"Deleted unannotated image: {file}")
            except Exception as e:
                logging.error(f"Failed to delete {file}: {e}")
        print(f"Deleted {len(unannotated_files)} unannotated images.")
    else:
        print("Deletion of unannotated images aborted by user.")
else:
    print("No unannotated images to remove.")
```

```python
annotation_counts = final_df_cleaned['file_id'].value_counts()
multiple_faces_files = annotation_counts[annotation_counts > 1].index.tolist()
print(f"Total images with multiple faces: {len(multiple_faces_files)}")
```

```python
if multiple_faces_files:
    pd.DataFrame({'file_id': multiple_faces_files}).to_csv(MULTIPLE_FACES_IMAGES_LOG, index=False)
    print(f"List of images with multiple faces saved to {MULTIPLE_FACES_IMAGES_LOG}")
    
    # **Safety Step**: Confirm before deletion
    confirmation = input(f"Are you sure you want to delete {len(multiple_faces_files)} images with multiple faces and their annotations? (yes/no): ")
    if confirmation.lower() == 'yes':
        # Remove annotations for these images
        annotations_to_remove = final_df_cleaned[final_df_cleaned['file_id'].isin(multiple_faces_files)]
        final_df_cleaned = final_df_cleaned[~final_df_cleaned['file_id'].isin(multiple_faces_files)].reset_index(drop=True)
        print(f"Removed {len(annotations_to_remove)} annotations for images with multiple faces.")
        
        # Remove the image files from the IMAGE_DIR
        for file in multiple_faces_files:
            file_path = os.path.join(IMAGE_DIR, file)
            try:
                os.remove(file_path)
                logging.info(f"Deleted image with multiple faces: {file}")
            except Exception as e:
                logging.error(f"Failed to delete {file}: {e}")
        print(f"Deleted {len(multiple_faces_files)} images with multiple faces.")
    else:
        print("Deletion of images with multiple faces aborted by user.")
else:
    print("No images with multiple faces found.")
```
